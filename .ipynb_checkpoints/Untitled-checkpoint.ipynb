```python
import matplotlib
import matplotlib.pyplot as plt
import pandas as pd
import datetime as dt
import seaborn as sns

%matplotlib inline
```

```python
# Change the style of the plots 
matplotlib.style.use('ggplot')
```

### Read dataset into pandas and some highlevel info about attributes

```python
# import the clean dataset
df_master = pd.read_csv('twitter_archive_master_final.csv')
df_master.info()
```

### Descriptive information for some attributes

```python
df_master[['rating_numerator','favorite_count','retweet_count']].describe()
```

- max rating_numerator is 1776. Since mean value is 12.23 , this seems to be an outlier


### Famous Dog Stages

```python
df_master['dog_stage'].value_counts().plot(kind = 'barh')
plt.title('Histogram of Most The Rated dog stages')
plt.xlabel('Count')
plt.ylabel('dog_breed')
```

As shown above , pupper is the most popular dog stages as per the data followed by doggo , puppo and floffer

### Popular dog names

```python
# Subseting the analysis for the names repeated atleast 5 times
df_by_name = df_master.groupby('name').filter(lambda x: len(x) > 4)

df_by_name['name'].value_counts().plot(kind = 'barh')
plt.title('Histogram of Most popular dog names')
plt.xlabel('Count')
plt.ylabel('dog_names')
```

Charlie is the Most common name as per the data

### Retweet count vs Favourite Count

```python
df_master.plot(kind = 'scatter', x = 'favorite_count', y = 'retweet_count', alpha = 0.5, color = 'red')
plt.xlabel('Favorites')
plt.ylabel('Retweets')
plt.title('Retweets and favorites Scatter plot')
```

As shown above there appears to be a correlation between retweets and favorite count

### Insights on rating_numerator

```python
# considering only valid values i.e. removing outliers
df_master[df_master['rating_numerator'] <= 14]['rating_numerator'].describe()
```

- More than 75% of the data has more than 12/10 as rating
- The page start with small rating than they adopt the system of rating numerator more than the denominator


### Dog stages with rating

```python
# Plot the dog stages with ratings
df_master[~ df_master['dog_stage'].isnull()].boxplot(column = ['rating_numerator'], by = ['dog_stage'])
plt.ylim(0,20)
plt.title('')
```

Puppers represent the big number of our pie, but it has the lowest mean rating
